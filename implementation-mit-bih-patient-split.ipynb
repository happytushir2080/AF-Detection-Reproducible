```python
bad = []
ok = 0
errors = 0

for p in mat_files:
    try:
        d = loadmat(p)
        if "ecg" not in d:
            errors += 1
            bad.append({"mat_path": os.path.relpath(p, DATA_ROOT), "reason": "missing_ecg"})
            continue

        x = np.asarray(d["ecg"]).squeeze()
        n = int(x.shape[0])

        # Prefer fs stored in mat; fallback to 250 if missing
        fs = float(np.asarray(d.get("fs", 250)).squeeze())
        dur = n / fs if fs > 0 else np.nan

        if dur < 10.0:
            # optional: label if present
            label = d.get("label", "")
            if isinstance(label, np.ndarray):
                try:
                    label = str(label.squeeze())
                except:
                    label = ""

            bad.append({
                "mat_path": os.path.relpath(p, DATA_ROOT),
                "num_samples": n,
                "fs": fs,
                "duration_sec": dur,
                "label": label
            })
        else:
            ok += 1

    except Exception as e:
        errors += 1
        bad.append({
            "mat_path": os.path.relpath(p, DATA_ROOT),
            "reason": f"load_error: {type(e).__name__}: {e}"
        })

bad_df = pd.DataFrame(bad)
print("OK (>=10s):", ok)
print("Bad (<10s or errors):", len(bad_df))
print("Errors:", errors)

bad_df.head(20)
```

```python
from scipy.io import loadmat
import numpy as np

def load_physionet_mat(mat_path: str):
    d = loadmat(mat_path)

    if "ecg" not in d:
        raise ValueError(f"Missing key 'ecg' in {mat_path}. Keys={list(d.keys())}")

    x = np.asarray(d["ecg"])

    # Make it 1D safely
    # Common shapes: (N,1), (1,N), (N,), sometimes (N,C)
    x = np.squeeze(x)

    # If still 2D (e.g., multichannel), take first channel
    if x.ndim == 2:
        # (N, C) or (C, N)
        if x.shape[0] > x.shape[1]:
            x = x[:, 0]
        else:
            x = x[0, :]

    x = np.asarray(x, dtype=np.float32).reshape(-1)  # force (N,)
    return x

from scipy.signal import filtfilt, lfilter

def _filtfilt_or_lfilter(b, a, x):
    x = np.asarray(x, dtype=np.float32).reshape(-1)  # FORCE 1D

    padlen = 3 * (max(len(a), len(b)) - 1)

    if x.size <= padlen:
        return lfilter(b, a, x).astype(np.float32)
    return filtfilt(b, a, x).astype(np.float32)
```

```python
def bandpass_filter(x, fs, low=0.5, high=40.0, order=4):
    nyq = 0.5 * fs
    b, a = butter(order, [low/nyq, high/nyq], btype="band")
    return _filtfilt_or_lfilter(b, a, x)

def notch_filter(x, fs, freq=50.0, Q=30.0):
    w0 = freq / (0.5 * fs)
    b, a = iirnotch(w0, Q)
    return _filtfilt_or_lfilter(b, a, x)


def ecg_log_spectrogram(x, fs, nperseg=64, noverlap=32, alpha=1e-6):
    """
    Returns log power spectrogram S: (F, T)
    Using torch.stft for consistency + speed.
    """
    xt = torch.tensor(x, dtype=torch.float32)

    window = torch.hann_window(nperseg)
    # stft: (freq, time, complex)
    Z = torch.stft(
        xt,
        n_fft=nperseg,
        hop_length=nperseg - noverlap,
        win_length=nperseg,
        window=window,
        center=False,
        return_complex=True
    )
    P = (Z.real**2 + Z.imag**2)  # power
    S = torch.log(P + alpha)
    return S.numpy().astype(np.float32)
```

```python
class MitBihAFSpecDataset(Dataset):
    def __init__(
        self,
        df: pd.DataFrame,
        input_dir: str,
        fs: int = 250,
        use_notch: bool = False,
        train: bool = False,
        spec_nperseg: int = 64,
        spec_noverlap: int = 32,
        spec_alpha: float = 1e-6,
    ):
        self.df = df.reset_index(drop=True)
        self.input_dir = input_dir
        self.fs = fs
        self.use_notch = use_notch
        self.train = train
        self.spec_nperseg = spec_nperseg
        self.spec_noverlap = spec_noverlap
        self.spec_alpha = spec_alpha

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):

        row = self.df.iloc[idx]
        seg_name = row["segment_name"]
        mat_rel  = row["mat_path"]
        label    = row["label"]
        y = 1.0 if label == "AF" else 0.0

        mat_path = os.path.join(self.input_dir, mat_rel)

        x = load_physionet_mat(mat_path)

        # DEBUG print if too short for filtfilt
        # (padlen for your butter order=4 bandpass is 27 in your error)
        if len(x) <= 27:
            print("\n[SHORT ECG] This will crash filtfilt")
            print("  segment_name:", seg_name)
            print("  mat_path:", mat_path)
            print("  label:", label)
            print("  len(x):", len(x))
            print("  first 20 samples:", x[:20])
            # raise to stop immediately with context
            raise ValueError(f"ECG too short for filtfilt: len={len(x)}")
            
        # row = self.df.iloc[idx]
        # seg_name = row["segment_name"]
        # mat_rel = row["mat_path"]
        # label = row["label"]

        # y = 1.0 if label == "AF" else 0.0

        # 1) load waveform
        # mat_path = os.path.join(self.input_dir, mat_rel)
        # x = load_physionet_mat(mat_path)

        # 2) filtering
        x = bandpass_filter(x, fs=self.fs, low=0.5, high=40.0, order=4)
        if self.use_notch:
            x = notch_filter(x, fs=self.fs, freq=50.0, Q=30.0)

        # 3) normalize waveform
        x = (x - x.mean()) / (x.std() + 1e-6)

        # 4) mild augmentation (train only)
        if self.train:
            x = x * np.random.uniform(0.9, 1.1)
            s = np.std(x)
            if s > 0:
                x = x + np.random.normal(0, 0.01 * s, size=x.shape).astype(np.float32)

        # 5) log-spectrogram
        S = ecg_log_spectrogram(
            x, fs=self.fs,
            nperseg=self.spec_nperseg,
            noverlap=self.spec_noverlap,
            alpha=self.spec_alpha
        )  # (F, T)

        # 6) normalize spectrogram
        S = (S - S.mean()) / (S.std() + 1e-6)

        return seg_name, torch.tensor(S, dtype=torch.float32), torch.tensor(y, dtype=torch.float32)

```

```python
def collate_pad(batch):
    # batch: list of (seg_name, S(F,T), y)
    names, specs, ys = zip(*batch)

    Fdim = specs[0].shape[0]
    T_max = max(s.shape[1] for s in specs)

    X = torch.zeros(len(batch), 1, Fdim, T_max, dtype=torch.float32)
    mask = torch.zeros(len(batch), T_max, dtype=torch.bool)
    y = torch.stack(ys).float()

    for i, S in enumerate(specs):
        T = S.shape[1]
        X[i, 0, :, :T] = S
        mask[i, :T] = True

    return list(names), X, mask, y
```

```python
import numpy as np
import pandas as pd

def patient_wise_split(df, seed=42, train_frac=0.7, val_frac=0.15, test_frac=0.15):
    assert abs(train_frac + val_frac + test_frac - 1.0) < 1e-9

    # Patient ID = record_name
    g = df.groupby("record_name")["label"].apply(list)

    # patient-level "has AF?"
    patient_has_af = g.apply(lambda labs: int(np.any(np.array(labs) == "AF")))
    patients = patient_has_af.index.to_numpy()
    y_pat = patient_has_af.values  # 1 if patient has any AF segment else 0

    # Stratified split at patient level
    rng = np.random.RandomState(seed)

    # split indices per class
    idx_pos = np.where(y_pat == 1)[0]
    idx_neg = np.where(y_pat == 0)[0]
    rng.shuffle(idx_pos)
    rng.shuffle(idx_neg)

    def split_indices(idxs):
        n = len(idxs)
        n_train = int(round(train_frac * n))
        n_val = int(round(val_frac * n))
        # remainder to test
        train_idx = idxs[:n_train]
        val_idx = idxs[n_train:n_train + n_val]
        test_idx = idxs[n_train + n_val:]
        return train_idx, val_idx, test_idx

    pos_tr, pos_va, pos_te = split_indices(idx_pos)
    neg_tr, neg_va, neg_te = split_indices(idx_neg)

    train_pat = set(patients[np.concatenate([pos_tr, neg_tr])])
    val_pat   = set(patients[np.concatenate([pos_va, neg_va])])
    test_pat  = set(patients[np.concatenate([pos_te, neg_te])])

    # Safety: no overlaps
    assert train_pat.isdisjoint(val_pat)
    assert train_pat.isdisjoint(test_pat)
    assert val_pat.isdisjoint(test_pat)

    train_df = df[df["record_name"].isin(train_pat)].reset_index(drop=True)
    val_df   = df[df["record_name"].isin(val_pat)].reset_index(drop=True)
    test_df  = df[df["record_name"].isin(test_pat)].reset_index(drop=True)

    return train_df, val_df, test_df, train_pat, val_pat, test_pat




```

```python
def masked_mean_pooling(x, mask):
    # x: (B,T,C)
    m = mask.unsqueeze(-1).float()
    x = x * m
    return x.sum(dim=1) / m.sum(dim=1).clamp(min=1.0)

def masked_topk_pooling(x, mask, k=5):
    # x: (B,T,C)
    neg_inf = torch.tensor(-1e9, device=x.device, dtype=x.dtype)
    x = torch.where(mask.unsqueeze(-1), x, neg_inf)
    k_eff = min(k, x.shape[1])
    topk_vals, _ = torch.topk(x, k=k_eff, dim=1)  # (B,k,C)
    return topk_vals.mean(dim=1)  # (B,C)




class SpecCRNNMaskedPoolNet(nn.Module):
    def __init__(self, dropout=0.3, lstm_hidden=64, feat_dim=32, topk=5):
        super().__init__()
        self.topk = topk

        self.conv = nn.Sequential(
            nn.Conv2d(1, 16, kernel_size=3, padding=1),
            nn.BatchNorm2d(16),
            nn.ReLU(),

            nn.Conv2d(16, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),

            nn.MaxPool2d(kernel_size=(2, 1)),  # 33 -> 16
            nn.Dropout(dropout),

            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),

            nn.MaxPool2d(kernel_size=(2, 1)),  # 16 -> 8
            nn.Dropout(dropout),
        )

        self.lstm = nn.LSTM(
            input_size=64 * 8,     # F' = 8
            hidden_size=lstm_hidden,
            num_layers=1,
            batch_first=True,
            bidirectional=True,
        )

        self.feat_drop = nn.Dropout(dropout)
        self.feat_fc = nn.Linear(2 * lstm_hidden * 2, feat_dim)  # mean+topk => 2*(2H)
        self.out_drop = nn.Dropout(dropout)
        self.out_fc = nn.Linear(feat_dim, 1)

    def forward(self, X, mask):
        z = self.conv(X)  # (B,64,F',T)
        B, C, Fp, T = z.shape

        # (B,64,F',T) -> (B,T,64*F')
        z = z.permute(0, 3, 1, 2).contiguous().view(B, T, C * Fp)

        out, _ = self.lstm(z)  # (B,T,2H)

        mean_pool = masked_mean_pooling(out, mask)
        topk_pool = masked_topk_pooling(out, mask, k=self.topk)
        pooled = torch.cat([mean_pool, topk_pool], dim=1)

        feat = F.relu(self.feat_fc(self.feat_drop(pooled)))
        logit = self.out_fc(self.out_drop(feat)).squeeze(-1)
        return logit


```

```python
def get_wave_and_spec(ds, idx):
    row = ds.df.iloc[idx]
    seg_name = row["segment_name"]
    mat_rel  = row["mat_path"]
    label    = row["label"]

    mat_path = os.path.join(ds.input_dir, mat_rel)
    x = load_physionet_mat(mat_path)

    # filtering
    x = bandpass_filter(x, fs=ds.fs, low=0.5, high=40.0, order=4)
    if ds.use_notch:
        x = notch_filter(x, fs=ds.fs, freq=50.0, Q=30.0)

    # normalize waveform
    x = (x - x.mean()) / (x.std() + 1e-6)

    # log-spectrogram
    S = ecg_log_spectrogram(
        x, fs=ds.fs, nperseg=ds.spec_nperseg,
        noverlap=ds.spec_noverlap, alpha=ds.spec_alpha
    )
    S = (S - S.mean()) / (S.std() + 1e-6)

    return seg_name, label, x, S


def plotly_wave_and_spec(seg_name, label, x, S, fs, split_name=""):
    t = np.arange(len(x)) / fs

    # Waveform
    fig_w = go.Figure()
    fig_w.add_trace(go.Scatter(x=t, y=x, mode="lines", name="ECG"))
    fig_w.update_layout(
        title=f"{split_name} | {seg_name} | label={label} | dur={t[-1]:.2f}s",
        xaxis_title="Time (s)",
        yaxis_title="Amplitude (z-scored)",
        height=350
    )
    fig_w.show()

    # Spectrogram heatmap: S is (F, T)
    fig_s = go.Figure(
        data=go.Heatmap(
            z=S,
            x=np.arange(S.shape[1]),
            y=np.arange(S.shape[0]),
            colorbar=dict(title="log-power (norm)")
        )
    )
    fig_s.update_layout(
        title=f"{split_name} | {seg_name} | Spectrogram (F x T) = {S.shape}",
        xaxis_title="Time bins",
        yaxis_title="Freq bins",
        height=450
    )
    fig_s.show()
```

```python
def eval_one_epoch(model, loader):
    model.eval()
    total_loss = 0.0
    n = 0
    logits_all = []
    y_all = []
    loss_fn = nn.BCEWithLogitsLoss()

    with torch.no_grad():
        for _, X, mask, y in loader:
            X = X.to(device, non_blocking=True)
            mask = mask.to(device, non_blocking=True)
            y = y.to(device, non_blocking=True)

            logit = model(X, mask)
            loss = loss_fn(logit, y)

            total_loss += loss.item() * y.size(0)
            n += y.size(0)

            logits_all.append(logit.detach().cpu())
            y_all.append(y.detach().cpu())

    logits_all = torch.cat(logits_all).numpy()
    y_all = torch.cat(y_all).numpy()

    probs = 1 / (1 + np.exp(-logits_all))
    pred = (probs >= 0.5).astype(np.int32)
    y_int = y_all.astype(np.int32)

    acc = (pred == y_int).mean()
    return total_loss / max(1, n), acc

def train_one_epoch(model, loader, optimizer, scaler=None):
    model.train()
    loss_fn = nn.BCEWithLogitsLoss()
    total_loss = 0.0
    n = 0

    for _, X, mask, y in loader:
        X = X.to(device, non_blocking=True)
        mask = mask.to(device, non_blocking=True)
        y = y.to(device, non_blocking=True)

        optimizer.zero_grad(set_to_none=True)

        if scaler is not None:
            with torch.cuda.amp.autocast():
                logit = model(X, mask)
                loss = loss_fn(logit, y)
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()
        else:
            logit = model(X, mask)
            loss = loss_fn(logit, y)
            loss.backward()
            optimizer.step()

        total_loss += loss.item() * y.size(0)
        n += y.size(0)

    return total_loss / max(1, n)
```

```python
import numpy as np
import torch
import pandas as pd

@torch.no_grad()
def get_probs_and_labels(model, loader, device):
    model.eval()
    probs_all = []
    y_all = []
    names_all = []

    for names, X, mask, y in loader:
        X = X.to(device, non_blocking=True)
        mask = mask.to(device, non_blocking=True)

        logit = model(X, mask)
        prob = torch.sigmoid(logit).detach().cpu().numpy()

        probs_all.append(prob)
        y_all.append(y.numpy())
        names_all.extend(names)

    probs_all = np.concatenate(probs_all).astype(np.float64)
    y_all = np.concatenate(y_all).astype(np.int32)
    names_all = np.array(names_all)

    return names_all, probs_all, y_all


```

```python
def compute_binary_stats(y_true, probs, thr=0.5, eps=1e-12):
    """
    y_true: 0/1
    probs:  predicted probability for class 1 (AF)
    """
    y_true = np.asarray(y_true).astype(np.int32)
    probs = np.asarray(probs).astype(np.float64)
    y_pred = (probs >= thr).astype(np.int32)

    TP = int(((y_pred == 1) & (y_true == 1)).sum())
    TN = int(((y_pred == 0) & (y_true == 0)).sum())
    FP = int(((y_pred == 1) & (y_true == 0)).sum())
    FN = int(((y_pred == 0) & (y_true == 1)).sum())

    # Core rates
    sensitivity = TP / (TP + FN + eps)      # TPR / Recall(AF)
    specificity = TN / (TN + FP + eps)      # TNR
    precision   = TP / (TP + FP + eps)      # PPV
    npv         = TN / (TN + FN + eps)      # NPV
    fpr         = FP / (FP + TN + eps)
    fnr         = FN / (FN + TP + eps)

    accuracy = (TP + TN) / (TP + TN + FP + FN + eps)
    balanced_acc = 0.5 * (sensitivity + specificity)

    f1 = 2 * precision * sensitivity / (precision + sensitivity + eps)

    # MCC (Matthews correlation coefficient)
    denom = np.sqrt((TP+FP)*(TP+FN)*(TN+FP)*(TN+FN) + eps)
    mcc = ((TP * TN) - (FP * FN)) / denom

    # Prevalence & predicted positive rate
    prevalence = (TP + FN) / (TP + TN + FP + FN + eps)
    pred_pos_rate = (TP + FP) / (TP + TN + FP + FN + eps)

    return {
        "threshold": float(thr),
        "TP": TP, "TN": TN, "FP": FP, "FN": FN,
        "accuracy": float(accuracy),
        "balanced_accuracy": float(balanced_acc),
        "sensitivity_recall_TPR": float(sensitivity),
        "specificity_TNR": float(specificity),
        "precision_PPV": float(precision),
        "NPV": float(npv),
        "F1": float(f1),
        "FPR": float(fpr),
        "FNR": float(fnr),
        "prevalence": float(prevalence),
        "pred_pos_rate": float(pred_pos_rate),
        "MCC": float(mcc),
    }
```

```python
def threshold_sweep_df(y_true, probs, split_name: str):
    rows = []

    thresholds = np.linspace(0.0001, 1.0000, 10000)  # 10,000 thresholds
    thresholds = np.round(thresholds, 4)  # clean 4-decimal precision

    for thr in thresholds:
        r = compute_binary_stats(y_true, probs, thr=float(thr))
        r["split"] = split_name
        rows.append(r)

    return pd.DataFrame(rows)

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve, average_precision_score

def plot_pr_curve(y_true, probs, label_name):
    y_true = np.asarray(y_true).astype(int)
    probs = np.asarray(probs).astype(float)

    precision, recall, thresholds = precision_recall_curve(y_true, probs)
    ap = average_precision_score(y_true, probs)

    plt.plot(recall, precision, label=f"{label_name} (AP={ap:.4f})")
    return ap

plt.figure(figsize=(7, 6))

ap_train = plot_pr_curve(y_train, probs_train, "Train")
ap_val   = plot_pr_curve(y_val,   probs_val,   "Val")
ap_test  = plot_pr_curve(y_test,  probs_test,  "Test")

# Baseline: prevalence (chance level for PR curve)
prev = y_test.mean()
plt.hlines(prev, 0, 1, linestyles="dashed", linewidth=1, label=f"Baseline (prev={prev:.3f})")

plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision–Recall Curves (AF as positive)")
plt.legend()
plt.grid(True)
plt.show()

print("Average Precision (AUPRC):")
print("  Train:", ap_train)
print("  Val:  ", ap_val)
print("  Test: ", ap_test)
```

```python
def plot_single_pr(y_true, probs, title):
    from sklearn.metrics import precision_recall_curve, average_precision_score
    p, r, _ = precision_recall_curve(y_true, probs)
    ap = average_precision_score(y_true, probs)

    plt.figure(figsize=(6,5))
    plt.plot(r, p)
    plt.xlabel("Recall")
    plt.ylabel("Precision")
    plt.title(f"{title} (AP={ap:.4f})")
    plt.grid(True)
    plt.show()
    

plot_single_pr(y_train, probs_train, "Train PR Curve")
plot_single_pr(y_val,   probs_val,   "Val PR Curve")
plot_single_pr(y_test,  probs_test,  "Test PR Curve")
```
